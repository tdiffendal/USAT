## Read in Employment Tables

```
#read in employment table 1
url <- 'https://www2.census.gov/programs-surveys/demo/tables/hhp/2020/wk9/employ1_week9.xlsx'
GET(url, mode = 'wb', write_disk(tf <- tempfile(fileext = ".xlsx")))

#create for loop to read in each sheet and save as df in list
n=1
employLoss = list()

#cycle through #sheets in xlsx
for (n in n:length(excel_sheets(tf))) {
  df <- read_xlsx(tf, n, skip=6, col_types=c('text','numeric','numeric','numeric',
  'numeric','numeric','numeric','numeric'))
  #ignore weirdly formatted first rows
  #df = df[6:nrow(df),] 
  #drop blank rows
  df = drop_na(df)
  #set column names
  colnames(df) <- c('characteristics','total','y_pastLoss','n_pastLoss',
  'na_pastLoss','y_futLoss','n_futLoss','na_futLoss')
  df['pct_y_pastLoss'] = (df['y_pastLoss'] * 100) / df['total']
  df['pct_n_pastLoss'] = df['n_pastLoss'] * 100 / df['total']
  df['pct_na_pastLoss'] = df['na_pastLoss'] * 100 / df['total']
  df['pct_y_futLoss'] = df['y_futLoss'] * 100 / df['total']
  df['pct_n_futLoss'] = df['n_futLoss'] * 100 / df['total']
  df['pct_na_futLoss'] = df['na_futLoss'] * 100 / df['total']
  #add df to list
  employLoss[[paste0(excel_sheets(path=tf)[n],"_employLoss")]]=df
}
```

```
#read in employment table 2
url <- 'https://www2.census.gov/programs-surveys/demo/tables/hhp/2020/wk9/employ2_week9.xlsx'
GET(url, write_disk(tf <- tempfile(fileext = ".xlsx")))

#create for loop to read in each sheet and save dfs to list
n=1
employType = list()

#cycle through #sheets in xlsx
for (n in n:length(excel_sheets(tf))) {
  df <- read_xlsx(tf, n)
  #ignore weirdly formatted first rows
  df <- df[7:nrow(df),] 
  #drop blank rows
  df <- drop_na(df)
  #set column names
  colnames(df) <- c('characteristics','total','employ_govt','employ_private',
  'employ_np','employ_self','employ_fam','employ_na', 'employ_no', 'na')
  #add df to list
  employType[[paste0(excel_sheets(path=tf)[n],"_employType")]]=df
}
```

## Read in Housing Tables

```
url = 'https://www2.census.gov/programs-surveys/demo/tables/hhp/2020/wk9/housing1a_week9.xlsx'
GET(url, write_disk(tf <- tempfile(fileext = ".xlsx")))

#create for loop to read in each sheet and save dfs to list
n=1
mortgage = list()

#cycle through #sheets in xlsx
for (n in n:length(excel_sheets(tf))) {
  df <- read_xlsx(tf, n)
  #ignore weirdly formatted first rows
  df <- df[6:nrow(df),] 
  #skip blank rows
  df <- drop_na(df)
  #set column names
  colnames(df) <- c('characteristics','total','own_house','n_mortgage',
  'y_mortgage','def_mortgage','na_mortgage','na')
  #add df to list
  mortgage[[paste0(excel_sheets(path=tf)[n],"_mortgage")]]=df
}
```
